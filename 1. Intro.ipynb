<span style="font-size:30px; font-weight:bold;">`Introduction`</span><br>
### `About the Dataset`

This project is based on Uber Eats USA Restaurants and Menus dataset. It has two main CSV files:

#### restaurants.csv
- contains details like restaurant name, rating, location, price range, etc.

#### restaurant-menus.csv
- contains menu items, their categories, prices, and descriptions.

### `Project Objective`
This project is based on restaurant and menu data from the U.S., where I’ve used SQL to find useful insights like state wise trends, popular food categories, and how restaurant ratings and performance vary from state to state. I’ve applied different SQL techniques like joins, subqueries, CTEs and window functions etc.
